```python
spares.loc[(spares['pup'] == '<Blank>'), 'pup'] = np.nan
```

```python
spares['pup'] = pd.to_datetime(spares.pup).astype('datetime64[s]')
```

```python
spares.loc[(spares['delivery_date_sanitised'] == '<Blank>'), 'delivery_date_sanitised'] = np.nan
spares['delivery_date_sanitised'] = pd.to_datetime(spares.delivery_date_sanitised).astype('datetime64[s]')
```

```python
from pandas.tseries.holiday import AbstractHolidayCalendar, Holiday, nearest_workday, \
    USMartinLutherKingJr, USPresidentsDay, GoodFriday, USMemorialDay, \
    USLaborDay, USThanksgivingDay
```

```python
class USCustomCalendar(AbstractHolidayCalendar):
    rules = [
        Holiday('NewYearsDay', month=1, day=1, observance=nearest_workday),
        USMartinLutherKingJr,
        USMemorialDay,
        Holiday('USIndependenceDay', month=7, day=4, observance=nearest_workday),
        USLaborDay,
        USThanksgivingDay,
        Holiday('Christmas', month=12, day=25, observance=nearest_workday)
    ]
```

```python
spares['lsp'] = 'UPS'
spares['department'] = 'Spares'
spares['shipped_date'] = spares.pup
spares['shipmentweight'] = 0
spares['origin_site'] = 'CVG'
spares['origin_cluster'] = 'CVG'
spares['origin_country'] = 'United States'
spares['origin_region'] = 'NALA'
spares['dep'] = pd.NaT
spares['arr'] = pd.NaT
spares['rdd'] = pd.NaT
spares['target_delivery_date'] = spares.apply(lambda x: x.pup + pd.offsets.CustomBusinessDay(n = 2 ), axis=1)
spares['estimated_delivery_date'] = spares['target_delivery_date'] 
spares['mot'] = 'GROUND'
spares['dbs_qdc'] = spares.qdc
spares['lane'] = 'None' 
spares['transit'] = (pd.to_datetime(spares['delivery_date_sanitised']) - pd.to_datetime(spares['pup'])).dt.days
spares['next_available_business_date'] = 'None'
spares['holiday_desc'] = 'None'
spares['delivery_type'] = 'DC Delivery'

spares['destination_cluster'] = 'None' # clean destination site
spares['destination_country'] = 'United States'
spares['destination_region'] = 'NALA'

spares['asset_id'] = spares.tracking_number
spares['time'] = 2

spares['sub_region'] = 'NALA-West'
spares['lat'] = 0
spares['lon'] = 0

spares['transit_time_sla'] = spares['transit']
spares['tdd_pup_sla'] = spares['target_delivery_date']
```

```python
spares['gross_sla_check'] = np.nan
spares.loc[spares['delivery_date_sanitised'] <= spares['target_delivery_date'], 'gross_sla_check'] = 1
```

```python
start = spares[spares['gross_sla_check'].isna()].pup.min()
end = spares[spares['gross_sla_check'].isna()].pup.max() + pd.Timedelta(days=30)
```

```python
for index, row in spares[spares['gross_sla_check'].isna()].iterrows():
    spares.at[index, 'target_delivery_date'] = row.pup + pd.offsets.CustomBusinessDay(n = 2, holidays = USCustomCalendar().holidays(start, end))
```

```python
spares.loc[pd.isna(spares.gross_sla_check) , 'gross_sla_check'] = 0
```

```python
spares = spares.merge(pod_reports_df[['Tracking#', 'Delay Reason (Select from drop down list)']], how = 'left', left_on = 'tracking_number', right_on = 'Tracking#')
```

```python
spares.qdc = spares['Delay Reason (Select from drop down list)']
spares.dbs_qdc = spares['Delay Reason (Select from drop down list)']
```

```python
spares.loc[spares['gross_sla_check'] == 0 & spares['qdc'].isna(), 'dbs_qdc'] = 'MISSING QDC'
spares.loc[spares['gross_sla_check'] == 0 & spares['qdc'].isna(), 'qdc'] = 'MISSING QDC'

spares.loc[spares['gross_sla_check'] == 0, 'controllable_flag'] = 'CONTROLLABLE'
spares.loc[(spares['gross_sla_check'] == 0) & (spares.qdc == "LATE AIRPLANE"), 'controllable_flag'] = 'CONTROLLABLE'
spares.loc[(spares['gross_sla_check'] == 0) & (spares.qdc == "BUSINESS CLOSED  1ST DELIVERY ATTEMPT"), 'controllable_flag'] = 'UNCONTROLLABLE'
```

```python
spares[spares.qdc.notna()].query('qdc!="MISSING QDC"')
```

```python
spares['net_sla_check'] = spares.gross_sla_check
```

```python
spares.loc[(spares['gross_sla_check'] == 0) & (spares.controllable_flag == "UNCONTROLLABLE"), 'net_sla_check'] = 1
```

```python
df = spares[['lsp','department','tracking_number','shipped_date','shipmentweight',
     'pup','dep','arr','rdd','estimated_delivery_date',
     'mot','delivery_date_sanitised','qdc','dbs_qdc','controllable_flag','target_delivery_date',
     'gross_sla_check','net_sla_check','lane','transit',
     'next_available_business_date','holiday_desc',
     'delivery_type','origin_site','destination_site','origin_cluster','destination_cluster',
     'origin_country','destination_country','origin_region','destination_region','asset_id','time', 'sub_region', 'lat', 'lon',	
        'transit_time_sla',	'tdd_pup_sla']]
```

```python
df
```

```python
df['gross_sla_check'] = df['gross_sla_check'].astype('Int32')  # Change to 'Int32' if INT is expected
df['net_sla_check'] = df['net_sla_check'].astype('Int32')
df.time = df.time.astype(str)
df.transit = df.transit.astype(float)
df.shipmentweight = df.shipmentweight.astype(float)
df.lat = df.lat.astype(float)
df.lon = df.lon.astype(float)
df.tracking_number = df.tracking_number.astype(str)
df.asset_id = df.asset_id.astype(str)
df.shipped_date = pd.to_datetime(df.shipped_date,errors='coerce').dt.date
df.pup = pd.to_datetime(df.pup,errors='coerce').dt.date
df.dep = pd.to_datetime(df.dep,errors='coerce').dt.date.astype('datetime64[s]')
df.arr = pd.to_datetime(df.arr,errors='coerce').dt.date.astype('datetime64[s]')
df.rdd = pd.to_datetime(df.rdd,errors='coerce').dt.date.astype('datetime64[s]')
df.estimated_delivery_date = pd.to_datetime(df.estimated_delivery_date,errors='coerce').dt.date
df.target_delivery_date = pd.to_datetime(df.target_delivery_date,errors='coerce').dt.date
df.delivery_date_sanitised = pd.to_datetime(df.delivery_date_sanitised,errors='coerce').dt.date

df['transit_time_sla'] = df['transit']
df['tdd_pup_sla'] = df['target_delivery_date']
```
